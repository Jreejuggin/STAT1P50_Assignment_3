```python
def get_frequency(input_string):
    list_of_words = input_string.split(' ')
    dict_of_frequencies = {}

    for word in list_of_words:
        if word in dict_of_frequencies.keys():
            dict_of_frequencies[word] = dict_of_frequencies[word] + 1
        else:
            dict_of_frequencies[word] = 1
    
    sorted_dict = dict(sorted(dict_of_frequencies.items(), key=lambda item: item[1])) 
    sortedList = list(sorted_dict.items())
    
    mostFreqWord = sortedList[len(sortedList)-1][1]
    
   
    freqRatio = mostFreqWord / len(list_of_words) 

    return(freqRatio)

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

data = pd.read_csv('email.csv')
listData = data['email']

# adding empty columns
data['hyperlink'] = None
data['free'] = None
data['click'] = None
data['business'] = None
data['frequency'] = None



#checks for each word in each email (each listData[i] is an email)
#if word is found puts 1 in column and if its not found puts 0 in column
for i in range (len(listData)):
    if ('hyperlink' in listData[i]) == True:
        data['hyperlink'][i] = 1
    else:
        data['hyperlink'][i] = 0
        
    if ('business' in listData[i]) == True:
        data['business'][i] = 1
    else:
        data['business'][i] = 0
        
    if ('free' in listData[i]) == True:
        data['free'][i] = 1
    else:
        data['free'][i] = 0
        
    if ('click' in listData[i]) == True:
        data['click'][i] = 1
    else:
        data['click'][i] = 0
        
    #gets frequency and puts it into the frequency column
    data['frequency'][i] = get_frequency(listData[i])


X = data[['hyperlink','business','free','click','frequency']] #features
y = data['label']  # Target variable 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print (accuracy)


    
```
